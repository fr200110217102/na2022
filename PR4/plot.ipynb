```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from math import pi,sin,exp
```

```python
A = [40 * (-sin(0.45*k*pi) + 2*sin(0.5*k*pi) - sin(0.55*k*pi)) / (k*k*pi*pi) for k in range(1,101)]
```

```python
def u(t, x):
    return sum([A[k-1] * exp(-k*k*pi*pi*t) * sin(k*pi*x) for k in range(1,101)])
```

```python
def plotall(T, files):
    x_arr = np.linspace(0, 1, 1000)
    plt.plot(x_arr, [u(T, x) for x in x_arr])
    for file in files:
        data = pd.read_csv(file)
        plt.plot(data['x'], data['u'])
    plt.show()
```

```python
def u(t, x):
    return exp(-20*(x-t-2)**2) + exp(-(x-t-5)**2)
```

```python
def plotall(T, files):
    x_arr = np.linspace(15, 25, 1000)
    plt.plot(x_arr, [u(T, x) for x in x_arr])
    for file in files:
        data = pd.read_csv(file)
        plt.plot(data[300:]['x'], data[300:]['u'])
    plt.show()
```
