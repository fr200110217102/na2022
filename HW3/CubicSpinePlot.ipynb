```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def f(x) :
    return 1 / (1 + 25*x**2)

x = np.linspace(-1, 1, 1000)
y = [f(t) for t in x]
plt.plot(x, y, '--')

file1 = pd.read_csv("a1.csv")
file2 = pd.read_csv("a2.csv")
file3 = pd.read_csv("a3.csv")
file4 = pd.read_csv("a4.csv")
file5 = pd.read_csv("a5.csv")

data = pd.DataFrame(file1)
plt.plot(data['x'], data['y'])
data = pd.DataFrame(file2)
plt.plot(data['x'], data['y'])
data = pd.DataFrame(file3)
plt.plot(data['x'], data['y'])
data = pd.DataFrame(file4)
plt.plot(data['x'], data['y'])
data = pd.DataFrame(file5)
plt.plot(data['x'], data['y'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def f(x) :
    return 1 / (1 + x**2)

x = np.linspace(-5, 5, 1000)
y = [f(t) for t in x]
plt.plot(x, y, '--')

file6 = pd.read_csv("b1.csv")
data = pd.DataFrame(file6)
plt.plot(data['x'], data['y'])

file7 = pd.read_csv("b2.csv")
data = pd.DataFrame(file7)
plt.plot(data['x'], data['y'])
```
