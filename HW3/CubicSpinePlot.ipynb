```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def f(x) :
    return 1 / (1 + 25*x**2)

x = np.linspace(-1, 1, 1000)
y = [f(t) for t in x]
plt.plot(x, y, '--')

file1 = pd.read_csv("s1.csv")
file2 = pd.read_csv("s2.csv")
file3 = pd.read_csv("s3.csv")
file4 = pd.read_csv("s4.csv")
file5 = pd.read_csv("s5.csv")

data = pd.DataFrame(file1)
plt.plot(data['x'], data['y'])
data = pd.DataFrame(file2)
plt.plot(data['x'], data['y'])
data = pd.DataFrame(file3)
plt.plot(data['x'], data['y'])
data = pd.DataFrame(file4)
plt.plot(data['x'], data['y'])
data = pd.DataFrame(file5)
plt.plot(data['x'], data['y'])
```
