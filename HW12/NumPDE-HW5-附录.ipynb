```python
import matplotlib.pyplot as plt
from numpy import *
from math import pi
```

```python
arr_mu = [0.8, 1.6, 2.0, 2.4]
m = 64
for mu in arr_mu:
    eigens = np.array([-mu/2 * (3 - 4*np.exp(-2j*pi*i/m) + np.exp(-4j*pi*i/m)) + mu**2/2 * (1 - 2*np.exp(-2j*pi*i/m) + np.exp(-4j*pi*i/m)) for i in range(m+1)])
    plt.scatter(eigens.real, eigens.imag, color = 'r')
    C = np.array([np.exp(1j*t)-1 for t in np.linspace(0, 2*pi, 1000)])
    plt.plot(C.real, C.imag)
    plt.show()
```
