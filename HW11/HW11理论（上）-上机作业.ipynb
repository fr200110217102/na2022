函数closedcurve：传入闭曲线解析式，作出闭曲线的图。

```python
theta = np.linspace(0, 2*pi, 1000)
def closedcurve(f):
    Z = np.exp(1j*theta)
    C = f(Z)
    plt.plot(C.real, C.imag)
```
